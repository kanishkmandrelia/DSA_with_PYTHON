SINGLY LINKED LIST

```python
class node:
    def __init__(self, item = None, next= None):
        self.item = item
        self.next = next
```

```python
class ssl:
    def __init__(self, start=None):
        self.start = start
    def is_empty(self):
        return self.start == None

        # Insertion #
    def insert_at_start(self, data):
        n = node(data,self.start)
        self.start = n
        self.print_list() 

    def insert_at_last(self,data):
        n = node(data)
        if not self.is_empty():
            p = self.start
            while (p.next!=None):
                p=p.next
            p.next = n
        else:
            self.start = n
        self.print_list()

    def insert_after(self,data,val):
        p = self.start
        if self.start == None:
            print("List is empty")
        else:
            while p.item != val:
                p = p.next
            n = node(data,p.next)
            p.next = n
        self.print_list()


        # Traversal #
    def print_list(self):
        p = self.start
        if self.start==None:
            print("List is empty")
        else:
            while(p!=None):
                print(p.item, end="->")
                p = p.next
            print("null")

    def search(self,val):
        p = self.start
        if p != None:
            while(p.item!=val):
                if p.next ==None:
                    print(f"the {val} is not present in the list")
                    break
                p= p.next
            if p.item==val:
                print(f"the {val} is present in the list")
        else:
            print("List is empty")
        

        # Deletion #
    def delete_first(self):
        if (self.start!=None):
            self.start = self.start.next
        else:
            print("List is empty")
        self.print_list()

    def delete_last(self):
        if self.start == None:
            print("List is empty\nnull")
        elif self.start.next == None:
            self.start = None
        else:
            p = self.start
            while(p.next.next!=None):
                p = p.next
            p.next = None
        self.print_list()

    def delete_item(self,data):
        if self.start == None:
            print("List is empty\nnull")
        elif self.start.next == None and self.start.item == data:
            self.start == None
        else:
            p = self.start
            if (p.item==data):
                self.start = p.next
                self.print_list()
            else:
                if p.next!= None:
                    while p.next.item !=data:
                        if p.next==None:
                            print(f"the {data} is not present in the list")
                            break
                        p = p.next
                    p.next = p.next.next
                else:
                    print("List is empty")
        self.print_list()
```

```python
l = ssl()
while True:
    choice = int(input("The choice:\n1=> Insert at start\n2=> Insert at last\n3=> Insert after particular value/data\n4=> Print the list\n5=> Search for the value/data\n6=> Delete first\n7=> Delete last\n8=> Delete particular data\n9=> Exit"))
    print("\n\nThe choice:\n1=> Insert at start\n2=> Insert at last\n3=> Insert after particular value/data\n4=> Print the list\n5=> Search for the value/data\n6=> Delete first\n7=> Delete last\n8=> Delete particular data\n9=> Exit\n")
    print("You have entered:\t",choice,"\n")
    if choice == 1:
        data = int(input("Enter the Data:"))
        print("The data you entered is:", data)
        l.insert_at_start(data)
    elif choice == 2:
        data = int(input("Enter the Data:"))
        print("The data you entered is:", data)
        l.insert_at_last(data)
    elif choice == 3:
        data = int(input("Enter the Data:"))
        print("The data you entered is:", data)
        val = int(input("Enter the value after which you want to enter the data:"))
        print(f"The data {data} is inserted after value {val}")
        l.insert_after(data,val)
    elif choice == 4:
        l.print_list()
    elif choice == 5:
        val = int(input("Enter the value which you want to search:"))
        print("The value which you have entered to search is:\t",val)
        l.search(val)
    elif choice == 6:
        l.delete_first()
    elif choice == 7:
        l.delete_last()
    elif choice == 8:
        data = int(input("Enter the data which you want to delete:"))
        print("The data you entered to delete is:\t",data)
        l.delete_item(data)
    elif choice == 9:
        break
    else:
        print("Please enter the valid number")
```
