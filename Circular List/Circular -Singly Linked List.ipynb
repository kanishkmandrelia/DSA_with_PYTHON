CIRCULAR (SINGLY) LINKED LIST

```python
class node:
    def __init__(self, item= None, next = None):
        self.item = item
        self.next = next
```

```python
class csll:
    def __init__(self, last =None):
        self.last = last


        # Insertion #
    def insert_at_first(self, data):
        if self.last == None:
            n= node(data)
            self.last = n
            n.next = n
        else:
            n = node(data,self.last.next)
            self.last.next = n
        self.print_list()

    def insert_at_last(self,data):
        if self.last == None:
            n = node(data)
            self.last = n
            n.next = n
        else:
            n = node(data,self.last.next)
            self.last.next= n
            self.last = n
        self.print_list()

    def insert_after(self,data,val):
        if self.last == None:
            raise IndexError("Circular Singly Linked List is empty")
        p = self.last.next
        while p.item != val:
            p = p.next
            if p == self.last.next:
                print(f"{val} is not present in the circular list.")
                break
        if p.item == val:
            n= node(data,p.next)
            p.next = n
            if p == self.last:
                self.last = n
        self.print_list()


        # Traversal #
    def search(self,data):
        if self.last != None:
            p  = self.last
            while p.item!=data:
                p = p.next
                if p == self.last:
                    print(data," is not present is the Circular Linked List")
                    break
            if p.item==data:
                print(data,"is present in the Circular Singly Linked list")
        else:
            print("Circular Singly Linked List is empty")
        self.print_list()

    def print_list(self):
        if self.last == None:
            print("Circular Singly Linked List is empty")
        else:
            p = self.last.next
            while p != self.last:
                print(p.item,end="->")
                p = p.next
            print(p.item,"->",p.next.item)
    

        # Deletion #
    def delete_first(self):
        if self.last == None:
            raise IndexError("Circular Singly Linked List is empty")
        elif self.last == self.last.next:
            self.last = None
        else:
            self.last.next = self.last.next.next
        self.print_list()

    def delete_last(self):
        if self.last !=None:
            if self.last == self.last.next:
                self.last = None
            else:
                p = self.last.next
                while p.next != self.last:
                    p = p.next
                p.next = self.last.next
                self.last = p
        else:
            raise IndexError("Circular Singly Linked List is empty")
        self.print_list()

    def delete_item(self,data):
        if self.last != None:
            if self.last == self.last.next:
                if self.last.item == data:
                    self.last=None
            else:
                if self.last.next.item == data:
                    self.delete_first()
                else:
                    p = self.last.next
                    while p!=self.last:
                        if self.last.item == data:
                            self.delete_last()
                            break
                        if p.next.item == data:
                            p.next = p.next.next
                            break
                        p  = p.next
            self.print_list()                
        else:
            raise IndexError("Circular Singly Linked List is empty")
```
