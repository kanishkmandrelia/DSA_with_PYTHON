DOUBLY - LINKED LIST

```python
class node:
    def __init__(self,  prev= None, item = None, next= None):
        self.prev = prev
        self.item = item
        self.next = next
```

```python
class dll:
    def __init__(self, start= None):
        self.start = start


            # Insertion #
        
    def insert_at_first(self,data):
        n= node(None,data,self.start)
        if self.start == None:
            self.start = n
        else:
            self.start.prev = n
            self.start = n
        self.print_list()

    def insert_at_last(self,data):
        if self.start!=None:
            p = self.start
            while p.next!= None:
                p = p.next
            n = node(p,data,None)
            p.next = n
        else:
            n = node(None,data,self.start)
            self.start = n
        self.print_list()

    def insert_after(self,data,val):
        p = self.start
        if self.start!= None:
            while p.item != val:
                p = p.next
                if p == None:
                    print(f"{val} is not present in the list.")
                    break
            n = node(p,data,p.next)
            if p.next == None:
                n = node(p,data,None)
            else:
                p.next.prev = n
            p.next = n
        else:
            print("Double Linked List is empty")
        self.print_list()


            # Traversal #

    def print_list(self):
        p = self.start
        while p!=None:
            print(p.item,end= "<=>")
            p =p.next
        print("null")

    def search(self,data):
        p = self.start
        if self.start!=None:
            while p.item != data:
                p= p.next
                if p.next == None and p.item !=data:
                    print(f"{data} is not in the list")
                    break
            if p.item==data:
                print(f"{data} is present in the list ")
        else:
            print("Double Linked List is empty")
        self.print_list()


            # Deletion #
                
    def delete_first(self):
        if self.start==None:
            print("List is empty")
        elif self.start.next == None:
            self.start =None
        else:
            p = self.start
            self.start = p.next
            p.next.prev = None
        self.print_list()

    def delete_last(self):
        if self.start==None:
            print("List is empty")
        elif self.start.next == None:
            self.start =None
        else:
            p = self.start
            while p.next!=None:
                p  = p.next
            p.prev.next= None
            p.prev = None
        self.print_list()
            
    def delete_between(self,data):
        p = self.start
        if self.start==None:
            print("List is empty")
        elif self.start.next == None:
            if self.start.item == data:
                self.start = None
            else:
                print(f"{data} is not present in the list")
        else:
            if self.start.item == data:
                self.start = p.next
                p.next.prev = None
            else:
                while p.next!=None:
                    if p.item != data:
                        p= p.next
                    else:
                        p.prev.next = p.next
                        p.next.prev = p.prev
                        break
                if p.item == data and p.next==None:
                    p.prev.next = None
                if p.next ==None and p.item!=data:
                    print(f"{data} is not present in the Doubly Linked List")
        self.print_list()
```
