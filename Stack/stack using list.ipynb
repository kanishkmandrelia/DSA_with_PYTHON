```python
class stack:
    def __init__(self):
        self.item = []
    def is_empty(self):
        if len(self.item) == 0:
            return "Stack is empty"
        
    def push(self,data):
        self.item.append(data)
        self.peek()
        
    def pop(self):
        if not self.is_empty():
            self.item.pop()
            self.peek()
        else:
            print("Stack is empty")
        
    def peek(self):
        if not self.is_empty():
            print("Top item: ",self.item[-1])
        else:
            print("Stack is empty")
        
    def size(self):
        print(len(self.item))
        
```

```python
s1 = stack()
while True:

    print("Enter the choice:\n1. Push the element\n2. Pop the element\n3. Top element of stack\n4. Size of stack\n5. Exit")
    choice = int(input("Enter the choice:\n1. Push the element\n2. Pop the element\n3. Top element of stack\n4. Size of stack\n5. Exit"))
    print("You have entered:\t", choice)

    if choice == 1:
        data = int(input("Enter the data to be inserted"))
        print("The data is: ", data)
        s1.push(data)
        print("\n")
    elif choice == 2:
        s1.pop()
        print("\n")
    elif choice == 3:
        s1.peek()
        print("\n")
    elif choice == 4:
        s1.size()
        print("\n")
    elif choice == 5:
        break
    else:
        print("Please enter the valid number")
        continue
```
